```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Fare"])
plt.title("Boxplot of Fare")
plt.show()
```

```python
#Observation
##Several outliers are visible in the Fare column.
#Most passengers paid fares within a lower price range.
##A few passengers paid significantly higher fares than the rest.
##These extreme values may influence statistical analysis and machine learning models.
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Age"])
plt.title("Boxplot of Age")
plt.show()
```

```python
#Observation
##The Age column contains a few outliers representing unusually old passengers.
##Most passengers fall within a moderate age range.
##The spread of the data indicates moderate variability in passenger ages.
```

```python
sns.pairplot(
    df,
    vars=["Age", "Fare", "Pclass"],
    hue="Survived"
)
plt.show()
```

```python
#Observation
##The pairplot shows relationships among Age, Fare, Passenger Class, and Survival.
##Passenger Fare tends to vary across different passenger classes.
##No strong linear relationship is observed between Age and Fare.
##A few isolated data points indicate the presence of outliers in the Fare feature.
##Most observations are concentrated in lower fare values
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr,
            annot=True,
            cmap="coolwarm")

```

```python
#Observation
##Most numerical features exhibit weak to moderate correlations.
##Passenger Class and Fare show a noticeable negative correlation, indicating that passengers in higher classes generally paid higher fares.
##SibSp and Parch display a slight positive correlation because families often traveled together.
##No pair of variables shows extremely high correlation, suggesting that severe multicollinearity is not present.
```

```python
sns.countplot(x="Survived", data=df)
```

```python
#Observation
##The number of passengers who did not survive is higher than those who survived.
##The dataset shows an imbalance between the two survival classes.
##Survival was less common than non-survival among passengers.
```
